## 3 Data Manipulation

### 3.1 Create data that provides dominant location of employees

```python
bottom_threshold = list(range(0,data_lokasi.shape[0],100))
up_threshold = list(range(99,data_lokasi.shape[0],100))
data_full = data_profil.copy()

for i,j in zip(bottom_threshold,up_threshold):
    one_date = pd.Series(data_lokasi.loc[i:j,'lokasi_dominan'],
                         name=str('date' + '_' + data_lokasi.loc[i,'tanggal'])).reset_index(drop=True)
    data_full = pd.concat([data_full,one_date],axis=1)
```

```python
data_full
```

### 3.2 Create data that provides dominant location of employees in boolean

```python
data_full_bool = data_full.copy()
```

```python
# List of cities in Jabodetabek
jabodetabek = ['Jakarta','Bogor','Depok','Tangerang','Bekasi']
```

```python
# Check if any cities in Jabodetabek
any(i in str('Jakarta Pusat').split() for i in jabodetabek)
```

```python
# Create manual function
def jabodetabek_bool(x):
    status = any(i in str(x).split() for i in jabodetabek)
    if status:
        return 1
    else:
        return 0
```

```python
for i in range(4,(data_full_bool.shape[1])):
    data_full_bool.iloc[:,i] = data_full_bool.iloc[:,i].apply(jabodetabek_bool)
```

```python
data_full_bool
```

```python
# Number of days in Jabodetabek
num_jabodetabek = data_full_bool.iloc[:,4:(data_full_bool.shape[1])].apply(sum,axis=1)
data_full_bool = pd.concat([data_full_bool,pd.Series(num_jabodetabek,name='number_jabodetabek')],axis=1)
```

```python
data_full_bool
```

```python
# Status 'mudik' = 1 or 'tidak mudik' = 0
stat_jabodetabek = [1 if i < 26 else 0 for i in data_full_bool['number_jabodetabek']]
data_full_bool = pd.concat([data_full_bool,pd.Series(stat_jabodetabek,name='status_jabodetabek')],axis=1)
```

```python
data_full_bool.head(50)
```

```python
data_transport = data_full_bool.filter(regex = ('date.*')).sum()
data_transport = data_transport.rename_axis('date').reset_index(name='freq')
```

```python
data_transport_lag = list(data_transport['freq'])
data_transport_lag.insert(0,0)
data_transport_lag.pop()
```

```python
data_transport_14 = pd.concat([data_transport,pd.Series(data_transport_lag,name='lag')],axis=1)
data_transport_14
```

```python
data_transport_14['num_mudik'] = data_transport_14['freq'] - data_transport_14['lag']
# Remove first and last obs
data_transport_14 = data_transport_14.drop([0,25])
data_transport_14
```

```python
data_transport_14.sort_values(by=['num_mudik'],ascending=True).head()
```
